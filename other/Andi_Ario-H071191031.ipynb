**Deskripsi dataset**

Dataset ini berisi 42.483 hasil pertandingan sepakbola internasional yang dimulai dari pertandingan resmi pertama sepakbola pada tahun 1972 hingga 2021. Data ini berisi pertandingan Piala Dunia dan pertandingan persahabatan. Pertandingan - pertandingan tersebut tidak termasuk pertandingan dalam Olimpiade dan pertandingan yang (minimal) salah satu timnya merupakan tim B, U-23, atau tim pilihan liga. Data ini pun hanya menyajikan pertandingan sepakbola pria, pertandingan sepakbola wanita tidak termasuk didalamnya.

Dalam dataset ini terdapat dua file csv, yaitu  results.csv dan  shootouts.csv.
Pada file results.csv berisi hasil pertandingan dalam waktu normal (Extra Time termasuk), sedangkan shootouts.csv berisi hasil pertandingan yang mesti diselesaikan lewat adu penalti.

Data pertandingan ini dikumpulkan dari beberapa sumber seperti Wikipedia, rsssf.com dan website asosiasi sepakbola serta beberapa sumber lainnya.

Beberapa value yang bisa didapat dari dataset ini:
1. Siapa tim terbaik sepanjang masa
2. Melihat goal terbanyak yang diciptakan suatu tim dalam satu pertandingan
3. Negara/ kota yang paling banyak menjadi tempat diselenggarakannya suatu pertandingan
4. Tim yang paling aktif
5. Tim yang mendominasi tiap era
6. Dan masih banyak lagi value yang dapat didapat dari statistik pertandingan yang tersedia

**Deskripsi Atribut**

1. `results.csv`

Pada file `results.csv` terdapat 9 kolom/atribut. Berikut penjelasan setiap atribut.

* `date`: berisi tanggal pertandingan
* `home_team`: nama dari tim kandang
* `away_team`: nama dari tim tandang
* `home_score`: skor akhir tim kandang (termasuk Extra Time)
* `away_score`: skor akhir tim tandang (termasuk Extra Time)
* `tournament`: nama dari turnamen terkait
* `city`: nama dari kota tempat pertandingan tersebut diselenggarakan
* `country`: nama dari negara tempat pertandingan tersebut diselenggarakan
* `neutral`: menyatakan apakah pertandingan tersebut diselenggarkan di tempat yang netral atau tidak (TRUE/FALSE)

2. `shootouts.csv`

Pada file `shootouts.csv` terdapat 4 kolom/atribut. Berikut penjelasan setiap atribut.

* `date`: berisi tanggal pertandingan
* `home_team`: nama dari tim kandang
* `away_team`: nama dari tim tandang
* `winner`: pemenang dari adu penalti
