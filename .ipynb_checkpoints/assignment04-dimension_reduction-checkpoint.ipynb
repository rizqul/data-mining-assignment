### handle using z-score

identified outliers by Z-score ( -3 > z & z > 3)

```python

from scipy import stats

z = np.abs(stats.zscore(df))
print(z)
#it is dificult to say which data point is an outlier

#defining threshold
threshold = 3
print(np.where(z > 3))
print(z)
```

```python
# before dump the outliers
df.shape
```

```python
# after dump the outliers
# use the Z-score

dataset_cleared_zscore = df[(z < 3).all(axis=1)]
dataset_cleared_zscore
```

### handle outliers using Algorithm for IQR (Interquartile Range)

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# before dump the outliers
df.shape
```

```python
# after dump the outliers
# using IQR

dataset_cleared_iqr = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(dataset_cleared_iqr.shape)
```

```python
df_out = dataset_cleared_iqr
```

## Resampling

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df_out.iloc[:,:-1]
y = df_out.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.2f%%' % (accuracy * 100.0))
```

```python
df['Treatment'].unique()
```

```python
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)
print('Confusion matrix:\n', conf_mat)

labels = ['Memantine', 'Saline']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```

```python
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
```

```python
print("Before : ",Counter(y))
```

1. Random Over Sampling

```python
# instantiating the random oversample
ros = RandomOverSampler()

# resampling X, y
X_ros, y_ros = ros.fit_resample(X, y)

# new class distribution
print("After : ",Counter(y_ros))
```

2. SMOTE (Synthetic Minority Oversampling Technique)

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

print(Counter(y_sm))
```

3. SMOTE Tomek

```python
from imblearn.combine import SMOTETomek

smt = smt = SMOTETomek(sampling_strategy='auto')
X_smt, y_smt = smt.fit_resample(X, y)
print(Counter(y_smt))
```

4. Random Under Sampling

```python
# instantiating the random undersampler
rus = RandomUnderSampler()

# resampling X, y
X_rus, y_rus = rus.fit_resample(X, y)

# new class distribution
print(Counter(y_rus))
```

5. Tomek Links

```python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks(sampling_strategy='majority')
X_tl, y_tl = tl.fit_resample(X, y)

print(Counter(y_tl))
```

6. cluster centroid

```python
from imblearn.under_sampling import ClusterCentroids

cc = ClusterCentroids()
x_cc, y_cc = cc.fit_resample(X, y)

print(Counter(y_cc))
```

3. SVM-RFE

```python
ranks = {}

def ranking(ranks, names, order=1):
  minmax = MinMaxScaler()
  ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
  ranks = map(lambda x: round(x,2), ranks)
  return dict(zip(names, ranks))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE

X = df_out.iloc[:, :-1]
y = df_out.iloc[:, -1]

M_train, n_train, M_test, n_test = train_test_split(X, y, test_size = 0.3, stratify = y)

colnames = X.columns
n_features_to_select = 1
svclassifier = SVC(kernel='linear')
rfe = RFE(svclassifier, n_features_to_select, verbose=1)
rfe.fit(X, y)

n_pred = rfe.predict(X)

ranks[f"RFE Score"] = ranking(list (map(float, rfe.ranking_)), colnames, order=1)
```

```python
df_rank = pd.DataFrame(list(ranks[f"RFE Score"].items()), columns=['Features', 'Scores'])

df_rank = df_rank.sort_values('Scores', ascending=False)
df_rank
```

## Feature Extraction

1. LDA

```python
from sklearn import manifold, datasets, decomposition, discriminant_analysis

lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=1).fit_transform(X, y)

lda_df = pd.DataFrame(data=lda,
                      columns=['new_var1'])
finalDf = pd.concat([lda_df, y], axis=1)
finalDf.head()
```

2. PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents,
                           columns=['principal_component1', 'principal_component2'])
finalDf=pd.concat([principalDf, y], axis=1)
finalDf.head()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)

X_2d = tsne.fit_transform(X)


new_df = pd.DataFrame(data=X_2d,
                      columns=['feature 1', 'feature 2'])
tane_df = pd.concat([new_df, y], axis=1)
tane_df.head()
```
